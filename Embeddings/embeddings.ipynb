```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY")
```

```python
# For converting text to vectors
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model = "text-embedding-3-large")
embeddings
```

```python
text = "This is to run a test case on OpenAi Embeddings"
q_results = embeddings.embed_query(text)

# Display the results

print(q_results)
```

```python
# shows 3072 as mentioned in documentation but we can change dimensions parameter
len(q_results)
```

```python
emb_1024 = OpenAIEmbeddings(model = "text-embedding-3-large", dimensions = 1024)
```

```python
text = "This is to run a test case on OpenAi Embeddings"
q_results = emb_1024.embed_query(text)

# Display the results

print(q_results)
len(q_results)
```

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("speech.txt")
docs = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 50)
fin_doc=text_splitter.split_documents(docs)
fin_doc
```

```python
# Vector embediings anf vector storeDB
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(fin_doc,emb_1024)
```

```python
# Retrieve the results from query vectorstore db
query="It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results=db.similarity_search(query)
print(retrieved_results)
```
